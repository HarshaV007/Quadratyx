```python
from collections import OrderedDict
import re
import pandas as pd


def kthRepeatedletter(filename,k):
    filename='InputText.txt'
    f = open(r"C:\Users\Harsha Vardhan\OneDrive\Desktop\QTX_Programming Question\InputText.txt")
    t=f.read()
    l=[]
    res = re.findall(r'\w+', t)
    for i in range(len(res)):
        input=res[i]
        dict=OrderedDict.fromkeys(input,0)  
        for ch in input:  
            
            dict[ch]+=1
        nonRepeatDict = [key for (key,value) in dict.items() if value==1]  
      
    # now return (k-1)th character from above list  
        if len(nonRepeatDict) < k:
            t='#'
        else:  
            t=nonRepeatDict[k-1]
        l.append(t)
    df = pd.DataFrame(data={"word":res,"character":l})
    df['k']=k
        
    return df
```

```python
def countCharacterType(filename): 
    filename='InputText.txt'
    f = open(r"C:\Users\Harsha Vardhan\OneDrive\Desktop\QTX_Programming Question\InputText.txt")
    t=f.read()
    l=[]
    res = re.findall(r'\w+', t)
    for i in range(len(res)):
        input=res[i]
    #Initiating values
        a = 0
        e = 0
        i = 0
        o = 0
        u = 0
        consonant = 0
 
        for n in range(0, len(input)):
            ch = input[n]  
            if ( (ch >= 'a' and ch <= 'z') or 
             (ch >= 'A' and ch <= 'Z') ):  
                ch = ch.lower() 
      
                if (ch == 'a'): 
                    a += 1
                elif(ch == 'e'):
                    e +=1
                elif(ch == 'i'):
                    i +=1 
                elif(ch == 'o'):
                    o +=1  
                elif(ch == 'u'):
                    u +=1    
                else: 
                    consonant += 1
          
        d={"Word":input,"A": a,"E": e,"I": i,"O": o,"U": u,"Consonants":consonant}
        l.append(d)
        df=pd.DataFrame.from_records(l)
    return df 
   
  
```
